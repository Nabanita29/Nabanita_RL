Co - Relation

```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
x = np.random.randint(0,50,1000)
```

```python
print(x)
```

```python
y = x + np.random.normal(0,10,1000)
```

```python
print(y)
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show()
```

```python
y2 = x - np.random.normal(0,10,1000)
```

```python
import matplotlib
matplotlib.style.use('ggplot')
plt.scatter(x,y2)
plt.show()
```

```python
y3 = 100 - x + np.random.normal(0,10,1000)
matplotlib.style.use('ggplot')
plt.scatter(x,y3)
plt.show()
```

```python
z = np.random.normal(0,30,1000)
y4 = z+np.random.normal(0,10,1000)
x2 = z+np.random.normal(0,50,1000)
```

```python
plt.scatter(x2,z)
plt.scatter(y4,z)
plt.show()
```

```python
np.corrcoef(x2,z)
```

```python
np.corrcoef(y4,z)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'a': np.random.normal(0,50,1000)})
df['b'] = df['a'] + np.random.normal(0,10,1000)
df['c'] = 100 - df['a'] + np.random.normal(0,5,1000)
df['d'] = np.random.randint(0,50,1000)
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show
```
